Since a boolean value of `True` is equivalent to `1` and a boolean value of `False` is equivalent to `0`, taking the mean of these values gives the proportion of `1`s. In this case, this is the proportion of values that are less than 0.5.

Not important, but here is the math behind this if it helps:

\begin{align}
\hat{\mu} &= \frac{1}{n}\sum_{i=1}^{n} x_i \\

\hat{\mu} &= \frac{1}{n} (0 + 0 + ... + 0 + 1 + 1 + ... + 1) \\

\hat{\mu} &= \frac{1}{n} (1 + 1 + ... + 1) \\ \\

\hat{\mu} &= \frac{1 + 1 + ... + 1}{n} \\
\end{align}
