# File Paths

During the day 2 session, attendees asked some amazing questions about file paths. While I (GK) answered some of these questions live, I think it would be helpful to have a resource here.

**Important note:** I will be using the forward slash `/` for file paths as opposed to the backslash `\`.

- On Windows you use backslashes (`\`)
- Pretty much everything else (Linux, MacOS, etc.) uses forward slashes (`/`)
    + So we will stick with forward slashes (`/`) for this document

## Things to know

- A folder is denoted by a slash after it
    + E.g., `folder-name/` (replace `folder-name` with the name of the folder)
- Folders do not have file extensions (e.g., a folder cannot be a `.pdf`)
    + E.g., `folder-name.pdf/` is incorrect
- Paths can be **relative** (important!) to the folder you are currently inside
    + E.g., this document is within the `q-and-a/` folder, so all paths will be relative to this folder

What do I mean by **relative**? Well, to access this file, the path would be `day2-attendee-q-and-a.ipynb`; that is, we are already inside the `q-and-a/` folder so we can already "see" this file. Thus, we do not write the folder name within the path.

The **root** can be defined as the outermost folder on your system. The location of this **root** folder differs by operating system.

- The GitHub repository also has a **root**. It is the outermost folder (i.e., the one containing the slides)
- Paths relative to the root are called **absolute paths**

Even though we do not, let us say that there is an imaginary sub-folder within `q-and-a/` called `workshop/`:

- If we are inside the `q-and-a/` folder, we can "see" the `workshop/` folder. Therefore, our (relative) path to it is simply `workshop/`
- Now, let's say we were not inside the `q-and-a/` folder, but instead at the **root** folder of the repository
    + The `workshop/` folder cannot be "seen" without first opening the `q-and-a/` folder
    + Therefore, the absolute path to the workshop folder is `q-and-a/workshop/`

Relative to the `q-and-a/` folder on the GitHub repository, how might we go "back" a folder so that we can "see" the root?:

- `../` means go to the 'parent' folder (the folder where your current sub-folder is contained)
- Therefore, if we are in `q-and-a/`, the path to our root folder is `../`
    + E.g, the relative path from the `q-and-a/` folder to the file `day1-slides.pdf` is `../day1-slides.pdf`
