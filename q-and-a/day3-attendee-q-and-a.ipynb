# Visualizing Groups in Scatterplots

There was a good question regarding how we can distinguish between groups within the same scatterplot as another way of assessing our data before statistical testing. Although we did end up seeing code that allowed us to do so towards the multiple regression section of this presentation, I wanted to go ahead and highlight how we brought this scatterplot to life here.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)   #makes the "random" data reproducible

#creating the intial dataframe with group and score information and storing it in df
df = pd.DataFrame({
    "group": ["control"] * 30 + ["treatment"] * 30,
    "score": np.concatenate([
        np.random.normal(70, 10, 30),
        np.random.normal(76, 10, 30)
    ])
})

#adding a study_hours column to df based on score values
df["study_hours"] = (
    2
    + 0.08 * df["score"]
    + np.random.normal(0, 1.5, len(df))
)

#filtering the dataframe according to group (treatment vs. control)
control_rows = df[df["group"] == "control"]
treatment_rows = df[df["group"] == "treatment"]

#this first plot layer adds the points from the control group rows
plt.plot(
    control_rows["study_hours"], control_rows["score"],
    "ob", label="Control")

#this second plot layer adds the points from the treatment group rows on top
plt.plot(
    treatment_rows["study_hours"], treatment_rows["score"],
    "or", label="Treatment")
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.legend() #pulls the marker format and label info from the .plot() layers
plt.show()
```

# Alternative Ways to Access Documentation

We talked about different ways of accessing documentation at the beginning to today's session, where I showed local alternatives to searching up the package and function of interest's documentation online. On that topic, I wanted to include a few resources that go over the three points of access I discussed: IntelliSense autocomplete within VS Code, appending `?`, and using the `help()` function.

- For Intellisense to work, you need to be working in Pythong environment, i.e. within a coding cell in a Jupyter notebook or in a terminal/interactive window based in a Python environment. Check out how this is supposed to look like on VS Code here: https://code.visualstudio.com/docs/languages/python#_autocomplete-and-intellisense.
- In order to use `?`, you should write out both the `package` and `function` names out before tacking it on at the end. This is because a bunch of different packages could have identically named functions that work in different ways, so the source package also needs to be identified. However, if there is no other function loaded into your workspace with that name, you can forego including the package name. This also means that in order to access the function of interest, you need to have already imported the source package.
- Similarly to the use of `?`, when getting documentation with the `help()` function, both the source package and function have to be specified and imported in.

```python
import matplotlib.pyplot as plt

#using ? and help() to get documentation
plt.plot?
help(plt.plot)

#example where an 'abs' function exists within both base and np packages
#notice the differences in documentation
np.abs?
abs?
```
