In this exercise, we will sort a list of integers using an algorithm called **bubble sort**. I will try and build the intuition gradually.

Below I construct an arbitrary list called `x`; I have designed it to require some sorting (i.e., from least to greatest):

```python
x = [1, 10, 3, 7, 4, 20, 2]
```

To avoid modifying the original list, we make a copy of it via the `copy` method. Let's name the variable `x_copy`:

```python
x_copy = x.copy()

# Note the use of underscores for variable naming as opposed to a 
# dot for calling a method or attribute.
```

Now, try to manually switch the second (`10`) and third element (`3`) of the list. You can do this through the use of a temporary variable, which you can name `temp`:

(Note that indices are equal to the element position minus one)

```python
temp = x_copy[1]
x_copy[1] = x_copy[2]
x_copy[2] = temp

print(x_copy)
```

Instead of using a temporary variable (explicitly), Python allows us to do this neatly in one line:

```python
x_copy[1], x_copy[2] = x_copy[2], x_copy[1]

print(x_copy) # x_copy is now the same as x again
```

Now we know how we can switch two elements in a list! This is the foundation for how we might sort it.

As it turns out, switching a pair of adjacent elements is what the **bubble sort** algorithm is all about.

For a first step to writing the algorithm, go ahead and check if the first two elements of the list are in the wrong order (i.e., greatest to least). If they are in the wrong order, switch them:

```python
if x_copy[0] > x_copy[0 + 1]: # Can also write 1 here instead of 0 + 1
        x_copy[0], x_copy[0 + 1] = x_copy[0 + 1], x_copy[1]

print(x_copy)
```

Now, we just need to figure out a way to generalize this process; namely, looping through the list, checking each pair of adjacent elements, and switching them if they are in the wrong order.

Let's first try using a `for` loop to do this. Below are a few hints:

- Before constructing the `for` loop, think about what the `range()` should be.
    + We are comparing element `i` to element `i + 1`
    + When `i` becomes the last element, what happens when we try to access `i + 1`?
- The list will not be sorted yet even if you have a correct `for` loop. We will address this later on.


```python
for i in range(len(x_copy) - 1): 
    if x_copy[i] > x_copy[i + 1]:
        x_copy[i], x_copy[i + 1] = x_copy[i + 1], x_copy[i]

print(x_copy)

# Note we use len(x_copy) - 1 because we are comparing i with i+1.
# If we used len(x_copy), when i becomes the index of the last element,
# i+1 will exceed the length of the list and return an error
```

The below lists the steps that the above `for` loop just took:

0. Check `1` and `3` --- do not switch
1. Check `3` and `10` --- do not switch
2. Check `10` and `7` --- switch

`x_copy` is now `[1, 3, 7, 10, 4, 20, 2]`

3. Check `10` and `4` --- switch

`x_copy` is now `[1, 3, 7, 4, 10, 20, 2]`

4. Check `10` and `20` --- do not switch
5. Check `20` and `2` --- switch

`x_copy` is now `[1, 3, 7, 4, 10, 2, 20]`

How about we take this `for` loop that we made and put it inside a `while` loop that continues to run until the list is fully sorted?

How might we use a counter variable to check if the list is fully sorted (and then exit the `while` loop)?

```python
count = 1       # So that we enter the loop at least once
while count > 0:
    count = 0   # Reset count to 0 at the beginning of each loop
    for i in range(len(x_copy) - 1):
        if x_copy[i] > x_copy[i + 1]:
            x_copy[i], x_copy[i + 1] = x_copy[i + 1], x_copy[i]
            count += 1 # Count the amount of times we switched elements
    # If count > 0, loop will go back to beginning

print(x_copy)
```

We can now see the intuition behind how the **bubble sort** algorithm works: the larger elements "bubble up" to the end of the list.

Let us now apply the algorithm to a new list of unsorted integers through the below steps:

1. Create a new list of unsorted integers called `y` 
2. Create `y_copy` via the `.copy` method 
3. Copy paste the code from the cell directly above (i.e., the one with the `while` loop) into the below cell
4. Replace all instances of `x_copy` with `y_copy`
5. Run the cell to see if `y_copy` is sorted

```python
y = [9, 9, 8, 8, 7, 7, 7]
y_copy = y.copy()

count = 1       # So that we enter the loop at least once
while count > 0:
    count = 0   # Reset count to 0 at the beginning of each loop
    for i in range(len(y_copy) - 1):
        if y_copy[i] > y_copy[i + 1]:
            y_copy[i], y_copy[i + 1] = y_copy[i + 1], y_copy[i]
            count += 1 # Count the amount of times we switched elements
    # If count > 0, loop will go back to beginning

print(y_copy)
```
