## Saving a converted type

During the day 1 session, the following code was used to convert the type of `x` from an `int` to a `str`:

```python
x = 416     # int
x = str(x)  # Convert type to str, save conversion

type(x)     # Check type of x
```

The question was (paraphrased), "Why do we have to save the conversion, `str(x)`, back into `x`?"

The answer:

- The function `str()` will convert its input into a `str`, but will not store it anywhere unless you tell it to. See below:

```python
x = 416     # int
str(x)      # Convert type to str, *do not save* conversion

type(x)     # Check type of x
```

## The `range()` function

"How do we set a starting point (other than the default of `0`) for the `range()` function?"

"How do we make the range function increment (e.g., `i`) by a number other than `1`?"

`range(start=0, stop, step=1)`

- `start` is starting point. Defaults to `0`
- `stop` is stopping point
- `step` is how much you increment by

```python
for i in range(2, 10, 3): # Start at 2, go up to (but excluding) 10, increment by 3 each time
    print(i)
```

# How to take slices in reverse-order

For example, if I have `x = [1,2,3,4,5]` and want to get a slice that is equal to `[4,3]` (i.e., [third index, second index]), how do we do this?

The first thing you need to know is that there is an extra component to slices that I left out of the slides (in an effort to keep things simple and more approachable):

Slices can take an additional 'stepping' parameter if you use another colon (`:`). As an example, let's say I want to take every second element (or index) of `x`:

```python
x = [1,2,3,4,5]

# start:stop:step

x[0:-1:2]   # 0 (inclusive) is first element, -1 (exclusive) is last element, 2 is the stepping parameter
```

If any of `start:stop:step` is not supplied (e.g., if you did not include a `start` value), Python defaults to the following:

- `start` is the beginning of the list (`0`)
- `stop` defaults to the end of the list; **however, the last element is included**
- `step` defaults to `1` (as seen in the slides)

```python
print(x[0:])    # From index 0 to end of list (output is the whole list)
print(x[:3])    # From the beginning of the list to index 3, excluding index 3
print(x[::2])   # See below
```

Note that for this last line, `x[::2]`, we take every second element of the list--but we start with the first element (zero-th index)

What happens if we do the same thing as in `x[::2]`, defaulting to the beginning of the list and the end of the list, but with a negative number instead (e.g., `x[::-1]`)?:

```python
x[::-1]
```

We get the reverse order!

For our last step, remember that we are trying to get the slice `[4,3]`; that is, a slice in format [third index, second index].

If the `step` parameter of our slice is a negative value, our `start` parameter should be a larger index than our `stop` parameter

- We are starting "ahead" in the list and then `step` backward until we `stop`:

```python
x[3:1:-1]   # Go from the third index to the first index, excluding the first index. Move backward through the list by a step of 1.
```
