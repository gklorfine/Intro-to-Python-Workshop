# Exercise 4: Understanding statsmodels Code & its Function Calls

The goal of this exercise is not to understand the statistics.

Instead, we're going to focus on identifying:
- Functions
- Arguments
- Objects

```python
import pandas as pd # importing the pandas package as `pd`
import statsmodels.formula.api as smf # importing the statsmodels package (specifically the formula.api module) as smf

np.random.seed(67)
score = np.random.normal(0, 1, 1000)
hours = np.random.normal(0, 30, 1000)

df = pd.DataFrame({'score': score, 'hours': hours})

# the fitted model is stored in `model`
model = smf.ols(            # smf.ols() creates the model
    "hours ~ score",        # the first argument provides the variables of interest in formula notation (outcome ~ predictor)
    data=df                 # the second argument provides the data to work with
).fit()                     # .fit() fits the created model

print(model.summary())      # provides the summary table with regression results
model.summary()             # does the same thing as print(model.summary())
```

# Exercise 6: Mini Analysis Pipeline

This exercise will combine plotting and modelling. Our goal is to complete each step in sequence and apply the general research workflow. Run the code below:

```python
import pandas as pd
import numpy as np

np.random.seed(42)

df = pd.DataFrame({
    "hours": np.random.uniform(0, 10, 50),
    "score": np.random.normal(75, 10, 50)
})
```

### Exercise 6.1

We are interested in understanding whether the number of hours spent studying significantly predict an individual's score (0% - 100%). Create a histogram of the `score` variable. Add:
- A title
- An x-axis label
- A y-axis label

```python
#Write your code for exercise 6.1 here:
# we use `plt` (a.k.a. the `matplotlib` package) to create our histogram
plt.hist(df["score"])
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Scores Histogram")
plt.show()
```

### Exercise 6.2

Get basic summary statistics like n and the mean for the hours and score variables. Create a scatterplot showing:
- hours on the x-axis
- score on the y-axis

Add:
- A title
- Axis labels

```python
#Write your code for exercise 6.2 here:
df.describe()                           # `.describe()` is a method that gets several basic descriptive stats

# we use `plt` (a.k.a. the `matplotlib` package) to create our scatterplot as well
plt.plot(df["hours"], df["score"], 'or')                # `.plot()` requiresx and y arguments to run
plt.title("Scatterplot of Exam Scores by Study Hours")
plt.xlabel("Hours spent studying")
plt.ylabel("Exam scores")
plt.show()
```

### Exercise 6.3

Let's fit a model where we use hours to predict scores and display its summary.

```python
#Write your code for exercise 6.3 here:
# we used different functions from the `statsmodels` package for our modelling
model = smf.ols("score ~ hours", data = df).fit() # `.fit()` is used to fit the model,
                                                  # no argument is needed, instead we chain the method onto our model

model.summary()                                   # .summary() is used to get summary results from our model
                                                  # again, no argument is needed, we just chain the method onto our model
```
